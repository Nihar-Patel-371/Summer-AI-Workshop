### Array Transformations

```python
# Element-wise operations
array_1d = np.array([1, 2, 3, 4, 5])
print("\nOriginal 1D Array:")
print(array_1d)

# Adding 10 to each element
array_transformed = array_1d + 10
print("\nArray after adding 10 to each element:")
print(array_transformed)

# Squaring each element
array_squared = array_1d ** 2
print("\nArray after squaring each element:")
print(array_squared)
```

### NumPy Reshaping

```python
# Reshaping a 1D array to 2D
array_1d = np.array([1, 2, 3, 4, 5, 6])
array_reshaped = array_1d.reshape((2, 3))
print("\nReshaped Array (2x3):")
print(array_reshaped)
print(array_reshaped.shape)

# Flattening a 2D array back to 1D
array_flattened = array_reshaped.flatten()
print("\nFlattened Array:")
print(array_flattened)
print(array_flattened.shape)
```

### Basic Array Operations

```python
# Basic arithmetic operations
array_a = np.array([1, 2, 3])
array_b = np.array([4, 5, 6])

print("\nArray A:")
print(array_a)
print("Array B:")
print(array_b)

# Addition
array_add = array_a + array_b
print("\nAddition of A and B:")
print(array_add)

# Multiplication
array_mul = array_a * array_b
print("\nMultiplication of A and B:")
print(array_mul)
print(array_a.dot(array_b))
```

```python

################################
#### Solution given below #####
################################

# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Check for null values
null_values = iris.isnull().sum()
print("Null values in each column:\n", null_values)

# There are no null values in the Iris dataset, but if there were:
# iris_cleaned = iris.dropna()

# Standardize the numerical features
scaler = StandardScaler()
numerical_features = iris.drop(columns=['species']).columns
iris_scaled = scaler.fit_transform(iris[numerical_features])

# Convert the scaled features back to a DataFrame
iris_scaled_df = pd.DataFrame(iris_scaled, columns=numerical_features)

# Add back the non-numerical columns
iris_final = pd.concat([iris_scaled_df, iris['species']], axis=1)

# Plot the data
sns.pairplot(iris_final, hue='species')

# Show the plot
plt.show()
```
